```python
import numpy as np
import gym
import random
from tensorflow.python.client import device_lib
print(device_lib.list_local_devices())
```

```python
from gym.envs.registration import register
register(
    id='FrozenLakeNotSlippery-v0',
    entry_point='gym.envs.toy_text:FrozenLakeEnv',
    kwargs={'map_name' : '8x8', 'is_slippery': False},
    max_episode_steps=100,
    reward_threshold=0.8196, # optimum = .8196, changing this seems have no influence
)
```

```python
env = gym.make("FrozenLake-v0")
env.reset()                    
env.render()
```

```python
action_size, state_size = env.action_space.n, env.observation_space.n
```

```python
qtable = np.zeros((state_size,action_size))
```

```python
total_episodes = 15000
total_test_episodes = 100
max_steps = 99
lr = 0.8
gamma = 0.95
epsilon = 1.0
min_epsilon = 0.01
max_epsilon = 1.0
decay_rate = 0.005
```

```python
rewards = []

for episode in range(total_episodes):
  
    state = env.reset()
    step = 0
    done = False
    total_rewards = 0

    for step in range(max_steps):
        
        random_num = random.uniform(0,1)

        if random_num > epsilon:
            action = np.argmax(qtable[state,:])
        else:
            action = env.action_space.sample()

        new_state, reward, done, info = env.step(action)
        qtable[state, action] = qtable[state, action] + lr * (reward + gamma * np.max(qtable[new_state, :]) - qtable[state, action])
        state = new_state
        total_rewards +=reward

        if done==True:
            break

    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)
    rewards.append(total_rewards)

print ("Score over time: " +  str(sum(rewards)/total_episodes))
print(qtable)
```

```python
for episode in range(total_test_episodes):
    state = env.reset()
    total_reward = 0
    step = 0
    done = False
  
    for step in range(max_steps):
        new_state, reward, done, info = env.step(action)
        total_reward += reward

        if done:
            env.render()
            break
        state = new_state
env.close()
print ("Score over time: " +  str(sum(rewards)/total_test_episodes))
```
